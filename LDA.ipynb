```python
import numpy as np
import math
import pandas as pd
import pyLDAvis
import pyLDAvis.lda_model
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Load your dataset
cleaned_data = pd.read_csv('preprocessed_cleaned_western_stereotypes.csv')

processedPosts = cleaned_data['Processed Text'].astype(str).tolist()

# Set parameters for CountVectorizer
featureNum = 1000  # Set this to desired number of features
topicNum = 20   # Adjust based on needs
wordNumToDisplay = 10  # Number of words to display per topic

# Create a CountVectorizer
tfVectorizer = CountVectorizer(max_df=0.95, min_df=2, max_features=featureNum, stop_words='english')
# Fit the vectorizer to the processed posts and transform into a matrix
tf = tfVectorizer.fit_transform(processedPosts)

# Fit the LDA model
ldaModel = LatentDirichletAllocation(n_components=topicNum, max_iter=10, learning_method='online', random_state=42)
ldaModel.fit(tf)

# Extract the names of the features
tfFeatureNames = tfVectorizer.get_feature_names_out()

# Function to display the topics
def display_topics(model, featureNames, numTopWords):
    """Prints out the most associated words for each topic."""
    for topicId, lTopicDist in enumerate(model.components_):
        print("Topic %d:" % (topicId))
        print(" ".join([featureNames[i] for i in lTopicDist.argsort()[:-numTopWords - 1:-1]]))

display_topics(ldaModel, tfFeatureNames, wordNumToDisplay)

# Visualize the topics using pyLDAvis
# Prepare the visualization panel for the LDA model
# panel = pyLDAvis.prepare(
#     ldaModel,
#     tf,
#     tfVectorizer,
#     vocab=tfFeatureNames,
#     term_frequency=tf.toarray().sum(axis=0)
# )

# Display the interactive visualization
# pyLDAvis.display(panel)

# Function to display word clouds
def displayWordcloud(model, featureNames):
    """Displays the word cloud of the topic distributions, stored in model."""
    # Normalize each row/topic to sum to one
    normalisedComponents = model.components_ / model.components_.sum(axis=1)[:, np.newaxis]
    
    topicNum = len(model.components_)
    plotColNum = 3  # Number of word clouds for each row
    plotRowNum = int(math.ceil(topicNum / plotColNum))

    plt.figure(figsize=(12, 8))
    for topicId, lTopicDist in enumerate(normalisedComponents):
        lWordProb = {featureNames[i]: wordProb for i, wordProb in enumerate(lTopicDist)}
        wordcloud = WordCloud(background_color='black', width=800, height=400)
        wordcloud.fit_words(frequencies=lWordProb)
        plt.subplot(plotRowNum, plotColNum, topicId + 1)
        plt.title('Topic %d:' % (topicId + 1))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis("off")

    plt.tight_layout()
    plt.show()

displayWordcloud(ldaModel, tfFeatureNames)
```
