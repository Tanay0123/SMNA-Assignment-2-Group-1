```python
import json
from collections import Counter
import matplotlib.pyplot as plt

# Load the YouTube data from the provided ab_data.json file
youtube_data_file = 'ab_data.json'

# Open and load the JSON file
with open(youtube_data_file, 'r', encoding='utf-8') as f:
    youtube_data = json.load(f)

# Initialize an empty list to store the channel names
channel_names = []

# List of terms to filter out from channel names
excluded_terms = ['shiseido', 'tirtir', 'sulwhasoo']

# Iterate through the dataset and collect the channel names
for video in youtube_data:
    # Extract the channel name and append it to the list
    channel_name = video.get('channel', None)
    if channel_name:
        # Check if any excluded term is in the channel name (case insensitive)
        if not any(term.lower() in channel_name.lower() for term in excluded_terms):
            channel_names.append(channel_name)

# Count the number of times each channel appears in the list
channel_count = Counter(channel_names)

# Display the most frequent channels
print("Frequent Channels and their Counts:")
for channel, count in channel_count.most_common():
    print(f"{channel}: {count}")

# Prepare data for the plot (top 10 most frequent channels)
channels, counts = zip(*channel_count.most_common(10))  # Get the top 10 most frequent channels

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.barh(channels, counts, color='skyblue')
plt.xlabel('Number of Videos')
plt.ylabel('Channel Name')
plt.title('Top 10 Most Frequent Channels Posting Content')
plt.gca().invert_yaxis()  # Invert y-axis to show the top channel at the top
plt.tight_layout()

# Show the plot
plt.show()
```
